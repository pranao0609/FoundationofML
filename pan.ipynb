```python
import pandas as pd
```

```python
#Series- Array Like 1D structure
data=[1,2,3,4,5]
series=pd.Series(data)
print(series)
```

```python
#Series using Dictionary
data ={'a':1 , 'b': 2, 'c':3}
series=pd.Series(data)
print(data)
```

```python
data=[10,20,30]
index=[1,2,3]
series=pd.Series(data,index=index)
print(series)
```

```python
#DataFrame

data={
    'Name':['Psa','May','Shr'],
    'Age': [19,20,20]
}
df=pd.DataFrame(data)
print(df)
```
