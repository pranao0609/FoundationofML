```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Basic Line Plot")
plt.plot(x,y)
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Basic Line Plot")
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```
